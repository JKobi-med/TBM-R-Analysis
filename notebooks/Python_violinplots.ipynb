# Violin plots with Python

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
Bdf = pd.read_excel('/home/kobi/Documents/02 Vocational/Promotionsdatananalyse mit Excel und Martin/Rohdaten/BDNF_Neuroplastizität_Alter_CpGs_ohne_m11_und_p36_mit_verhalten_vollständig_umstrukturiert.xlsx')

Tdf = pd.read_excel('/home/kobi/Documents/02 Vocational/Promotionsdatananalyse mit Excel und Martin/Rohdaten/TAU_CpGs_mit_verhalten_ohne_p61_p59_p34_p29_umstruk.xlsx')

Pdf = pd.read_excel("/home/kobi/Documents/02 Vocational/Promotionsdatananalyse mit Excel und Martin/Rohdaten/PSD95_CpGs_mit_verhalten_größer_umstrukturiert.xlsx")
```

```python
type(Bdf)
```

## violinplots

```python
plt.figure(figsize = [10,5])
sns.violinplot(data = Bdf, x = "CpG_position", y = 'methylierung', 
               hue = 'Gruppe', split = True)
plt.title('BDNF CpG methylation by Group')
plt.xlabel('CpG-Position')
plt.ylabel('Methylation-Values')
plt.savefig('BDNF_CpG_methylation_by_group.png', dpi = 300)
```

```python
plt.figure(figsize = [25,5])
sns.violinplot(data = Tdf, x = "CpG_position", y = "methylation", 
               hue = "Gruppe", split = True)
plt.title('Tau CpG methylation by Group')
plt.xlabel('CpG-Position')
plt.ylabel('Tau-Methylation-Values')
plt.savefig('Tau_CpG_methylation_by_group.png', dpi = 300)
```

```python
plt.figure(figsize = [24,5])
sns.violinplot(data = Pdf, x = "CpG_Position", y = "methylation", 
               hue = "Gruppe", split = True)
plt.title('Tau CpG methylation by Group')
plt.xlabel('CpG-Position')
plt.ylabel('PSD95-Methylation-Values')
plt.savefig('PSD95_CpG_methylation_by_group.png', dpi = 300)
```
